```python
from openai import OpenAI

client = OpenAI(api_key = "sk-proj")
```

```python
# Create a vector store caled "Financial Statements"
vector_store = client.vector_stores.create(name="Financial Statements")

# Ready the files for upload to OpenAI
file_paths = ["goog-10k.pdf", "brka-10k.pdf"]
file_streams = [open(path, "rb") for path in file_paths]

# Use the upload and poll SDK helper to upload the files, add them to the vector store,
# and poll the status of the file batch for completion.
file_batch = client.vector_stores.file_batches.upload_and_poll(
  vector_store_id=vector_store.id, files=file_streams
)

# print the status and the file counts of the batch to see the result of this operation.
print(file_batch.status)
print(file_batch.file_counts)
```

```python
# Import the OpenAI library
from openai import OpenAI

# Create an OpenAI client instance
client = OpenAI()

# Create a response using the GPT-4o-mini model
response = client.responses.create(
    model="gpt-4o-mini", 
    input="Can you summarize the main risks Google reported in its 10-K?",  # User query related to financial analysis
    instructions="You are an expert financial analyst. Use your knowledge base to answer questions about audited financial statements.",  
    # Instructions to guide the assistant's behavior—acts as a prompt to shape the expert tone and scope
    tools=[{
        "type": "file_search",  # Use file search tool to retrieve relevant documents
        "vector_store_ids": [vector_store.id],  # Link to a specific vector store that likely contains indexed 10-K documents
    }]
)

# Print the output text generated by the model
print(response.output_text)
```

## METADATA FILTERING

```python
# Import the OpenAI library
from openai import OpenAI

# Create a client instance to interact with OpenAI API
client = OpenAI()

# Create a response by sending a request to the GPT model
response = client.responses.create(
    model="gpt-4o-mini",  
    input="Can you summarize the main risks Google reported in its 10-K?",  # Question about Google's risk disclosures in its 10-K filing
    instructions="You are an expert financial analyst. Use your knowledge base to answer questions about audited financial statements.",
    # Instructions to guide the tone and depth of analysis expected from the model
    
    tools=[{
        "type": "file_search",  # Enable the use of a file search tool for document retrieval
        "vector_store_ids": [vector_store.id],  # Use a specific vector store that contains relevant financial documents

        # Apply a filter to narrow down the search results to only 10-K reports
        "filters": {
            "type": "eq",          # Type of comparison: equality check
            "key": "report_type",  # Metadata field to filter on
            "value": "10-K"        # Only include documents where report_type == "10-K"
        }
    }]
)

# Print the model's summarized output from the filtered 10-K document(s)
print(response.output_text)
```
